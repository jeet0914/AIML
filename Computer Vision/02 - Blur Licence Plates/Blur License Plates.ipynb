# Use Haar Cascades to identify and blur license plates

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/car_plate.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def display(img):
    
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display(img)
```

```python
display(img)
```

```python
har_russ = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    
    car_plate = img.copy()
    
    detect_num_plate = har_russ.detectMultiScale(car_plate, scaleFactor = 1.2, minNeighbors = 3)
    
    for x,y,w,h in detect_num_plate:
        
        cv2.rectangle(car_plate, (x,y), (x+w, y+h), 255, 3)
        
    return car_plate
    
    
    
```

```python
result = detect_plate(img)
```

```python
display(result)
```

```python
display(result)
```

```python
def detect_and_blur_plate(img):
    
    car_plate = img.copy()
    
    detect_num_plate = har_russ.detectMultiScale(car_plate, scaleFactor = 1.2, minNeighbors = 3)
        
    for x,y,w,h in detect_num_plate:
    
    
        car_plate[y:y+h, x:x+w] = cv2.medianBlur(car_plate[y:y+h, x:x+w], 7)
    
    return car_plate
    
```

```python
display(detect_plate(img))
```

```python
print(detect_and_blur_plate(img))
```

```python
result = detect_and_blur_plate(img)
```

```python
display(result)
```
